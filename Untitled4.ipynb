```python
import pandas as pd
import numpy as np

def run_adr_signal_detection():
    """
    Main function to run the ADR Signal Detection analysis.
    """
    try:
        # --- Step 1: Load the Dataset ---
        print("Step 1: Loading dataset 'medicine_dataset.csv'...")
        # Use error_bad_lines=False to skip problematic rows if any
        df = pd.read_csv('medicine_dataset.csv', on_bad_lines='skip')
        print("Dataset loaded successfully.\n")

        # --- Step 2: Preprocess and Clean the Data ---
        print("Step 2: Preprocessing and cleaning data...")
        
        # Identify all 'sideEffect' columns
        side_effect_cols = [col for col in df.columns if col.startswith('sideEffect')]
        
        # Use pd.melt() to "unpivot" the side effect columns
        # This changes the data from "wide" to "long" format
        df_melted = df.melt(
            id_vars=['name'], 
            value_vars=side_effect_cols, 
            value_name='Side Effect'
        )
        
        # Clean up the resulting dataframe
        # 1. Drop the 'variable' column (we don't need 'sideEffect0', 'sideEffect1', etc.)
        df_melted = df_melted.drop(columns=['variable'])
        
        # 2. Clean the 'Side Effect' text data
        # Convert to string, strip whitespace, and convert to lowercase
        df_melted['Side Effect'] = df_melted['Side Effect'].astype(str).str.strip().str.lower()
        
        # 3. Drop rows where 'Side Effect' is missing (NaN) or is a placeholder like 'nan'
        df_melted = df_melted.dropna(subset=['Side Effect'])
        df_melted = df_melted[df_melted['Side Effect'] != 'nan']
        
        # 4. Clean the 'name' (drug name) column
        df_melted['name'] = df_melted['name'].astype(str).str.strip().str.lower()
        
        print("Data preprocessing complete. We now have one row per reported drug-side effect.\n")

        # --- Step 3: Perform Frequency Analysis ---
        print("--- Step 3: Frequency Analysis ---")
        
        # 3a. Overall Top 20 Most Common Side Effects (All Drugs)
        top_overall_effects = df_melted['Side Effect'].value_counts().head(20)
        print("\n[Analysis] Top 20 Most Common Side Effects (Overall):")
        print(top_overall_effects)
        print("-" * 50)
        
        # 3b. Top 20 Most Common Drug-Side Effect Pairs
        # This gives us a basic idea of strong associations
        df_pairs = df_melted.groupby(['name', 'Side Effect']).size().reset_index(name='Count')
        df_pairs = df_pairs.sort_values(by='Count', ascending=False)
        
        print("\n[Analysis] Top 20 Most Frequent Drug-Side Effect Pairs:")
        print(df_pairs.head(20))
        print("-" * 50)

        # --- Step 4: Calculate Statistical Signal (Proportional Reporting Ratio - PRR) ---
        print("\n--- Step 4: Calculating Proportional Reporting Ratio (PRR) ---")
        # PRR is a statistical measure of disproportionality.
        # PRR = (a / (a+c)) / (b / (b+d))
        # Where, for a given Drug X and Side Effect Y:
        # a = Count of (Drug X, Side Effect Y)
        # b = Count of (Other Drugs, Side Effect Y)
        # a+c = Total reports for Drug X
        # b+d = Total reports for Other Drugs
        
        # Calculate N (Total number of all reports)
        N_total_reports = df_pairs['Count'].sum()
        
        # Calculate a+c (Total reports for each drug)
        drug_totals = df_pairs.groupby('name')['Count'].sum().reset_index(name='a_plus_c')
        
        # Calculate a+b (Total reports for each side effect)
        effect_totals = df_pairs.groupby('Side Effect')['Count'].sum().reset_index(name='a_plus_b')
        
        # Merge these stats back into our pairs dataframe
        df_signals = pd.merge(df_pairs, drug_totals, on='name')
        df_signals = pd.merge(df_signals, effect_totals, on='Side Effect')
        
        # Now we can calculate all components of the 2x2 table
        df_signals['a'] = df_signals['Count']
        df_signals['a_plus_c'] = df_signals['a_plus_c']
        df_signals['a_plus_b'] = df_signals['a_plus_b']
        df_signals['b'] = df_signals['a_plus_b'] - df_signals['a']
        df_signals['b_plus_d'] = N_total_reports - df_signals['a_plus_c']

        # Calculate the two incidences
        # Incidence in our drug of interest
        df_signals['incidence_drug'] = df_signals['a'] / df_signals['a_plus_c']
        
        # Incidence in all other drugs (handle division by zero)
        df_signals['incidence_other'] = np.where(
            df_signals['b_plus_d'] > 0,
            df_signals['b'] / df_signals['b_plus_d'],
            0
        )
        
        # Calculate PRR (handle division by zero)
        # If incidence_other is 0, PRR is infinite (strong signal)
        df_signals['PRR'] = np.where(
            df_signals['incidence_other'] > 0,
            df_signals['incidence_drug'] / df_signals['incidence_other'],
            np.inf
        )
        
        print("PRR calculation complete.\n")

        # --- Step 5: Identify and Display Potential Signals ---
        print("--- Step 5: Identifying Potential ADR Signals ---")
        # A common threshold for a "signal" is:
        # 1. PRR > 2 (The side effect is reported >2x as often for this drug)
        # 2. Count (a) >= 3 (The association is not just a random one-off report)
        
        potential_signals = df_signals[
            (df_signals['PRR'] > 2) & 
            (df_signals['a'] >= 3)
        ].sort_values(by='PRR', ascending=False)
        
        print(f"\nFound {len(potential_signals)} potential signals (PRR > 2 and Count >= 3).")
        print("\nTop 20 Potential Signals (Sorted by PRR):")
        
        # Select key columns for a cleaner output
        output_cols = ['name', 'Side Effect', 'Count', 'PRR', 'a_plus_c', 'a_plus_b']
        print(potential_signals[output_cols].head(20).to_string())
        
        print("\n\n--- ADR Signal Detection Project Complete ---")
        print("Interpretation:")
        print(" - 'Count' (a): Number of reports for this specific drug-effect pair.")
        print(" - 'PRR': Proportional Reporting Ratio. A value of 5.0 means this side effect is reported 5x more frequently for this drug than for all other drugs in the dataset.")
        print(" - 'a_plus_c': Total side effect reports for this drug.")
        print(" - 'a_plus_b': Total reports for this side effect across all drugs.")

    except FileNotFoundError:
        print(f"ERROR: The file 'medicine_dataset.csv' was not found.")
        print("Please make sure the file is in the same directory as the script.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    run_adr_signal_detection()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

def run_adr_signal_detection():
    """
    Main function to run the ADR Signal Detection analysis.
    """
    try:
        # --- Step 1: Load the Dataset ---
        print("Step 1: Loading dataset 'medicine_dataset.csv'...")
        # Use error_bad_lines=False to skip problematic rows if any
        df = pd.read_csv('medicine_dataset.csv', on_bad_lines='skip')
        print("Dataset loaded successfully.\n")

        # --- Step 2: Preprocess and Clean the Data ---
        print("Step 2: Preprocessing and cleaning data...")
        
        # Identify all 'sideEffect' columns
        side_effect_cols = [col for col in df.columns if col.startswith('sideEffect')]
        
        # Use pd.melt() to "unpivot" the side effect columns
        # This changes the data from "wide" to "long" format
        df_melted = df.melt(
            id_vars=['name'], 
            value_vars=side_effect_cols, 
            value_name='Side Effect'
        )
        
        # Clean up the resulting dataframe
        # 1. Drop the 'variable' column (we don't need 'sideEffect0', 'sideEffect1', etc.)
        df_melted = df_melted.drop(columns=['variable'])
        
        # 2. Clean the 'Side Effect' text data
        # Convert to string, strip whitespace, and convert to lowercase
        df_melted['Side Effect'] = df_melted['Side Effect'].astype(str).str.strip().str.lower()
        
        # 3. Drop rows where 'Side Effect' is missing (NaN) or is a placeholder like 'nan'
        df_melted = df_melted.dropna(subset=['Side Effect'])
        df_melted = df_melted[df_melted['Side Effect'] != 'nan']
        
        # 4. Clean the 'name' (drug name) column
        df_melted['name'] = df_melted['name'].astype(str).str.strip().str.lower()
        
        print("Data preprocessing complete. We now have one row per reported drug-side effect.\n")

        # --- Step 3: Perform Frequency Analysis ---
        print("--- Step 3: Frequency Analysis ---")
        
        # 3a. Overall Top 20 Most Common Side Effects (All Drugs)
        top_overall_effects = df_melted['Side Effect'].value_counts().head(20)
        print("\n[Analysis] Top 20 Most Common Side Effects (Overall):")
        print(top_overall_effects)
        print("-" * 50)
        
        # 3b. Top 20 Most Common Drug-Side Effect Pairs
        # This gives us a basic idea of strong associations
        df_pairs = df_melted.groupby(['name', 'Side Effect']).size().reset_index(name='Count')
        df_pairs = df_pairs.sort_values(by='Count', ascending=False)
        
        print("\n[Analysis] Top 20 Most Frequent Drug-Side Effect Pairs:")
        print(df_pairs.head(20))
        print("-" * 50)

        # --- Step 4: Calculate Statistical Signal (Proportional Reporting Ratio - PRR) ---
        print("\n--- Step 4: Calculating Proportional Reporting Ratio (PRR) ---")
        # PRR is a statistical measure of disproportionality.
        # PRR = (a / (a+c)) / (b / (b+d))
        # Where, for a given Drug X and Side Effect Y:
        # a = Count of (Drug X, Side Effect Y)
        # b = Count of (Other Drugs, Side Effect Y)
        # a+c = Total reports for Drug X
        # b+d = Total reports for Other Drugs
        
        # Calculate N (Total number of all reports)
        N_total_reports = df_pairs['Count'].sum()
        
        # Calculate a+c (Total reports for each drug)
        drug_totals = df_pairs.groupby('name')['Count'].sum().reset_index(name='a_plus_c')
        
        # Calculate a+b (Total reports for each side effect)
        effect_totals = df_pairs.groupby('Side Effect')['Count'].sum().reset_index(name='a_plus_b')
        
        # Merge these stats back into our pairs dataframe
        df_signals = pd.merge(df_pairs, drug_totals, on='name')
        df_signals = pd.merge(df_signals, effect_totals, on='Side Effect')
        
        # Now we can calculate all components of the 2x2 table
        df_signals['a'] = df_signals['Count']
        df_signals['a_plus_c'] = df_signals['a_plus_c']
        df_signals['a_plus_b'] = df_signals['a_plus_b']
        df_signals['b'] = df_signals['a_plus_b'] - df_signals['a']
        df_signals['b_plus_d'] = N_total_reports - df_signals['a_plus_c']

        # Calculate the two incidences
        # Incidence in our drug of interest
        df_signals['incidence_drug'] = df_signals['a'] / df_signals['a_plus_c']
        
        # Incidence in all other drugs (handle division by zero)
        df_signals['incidence_other'] = np.where(
            df_signals['b_plus_d'] > 0,
            df_signals['b'] / df_signals['b_plus_d'],
            0
        )
        
        # Calculate PRR (handle division by zero)
        # If incidence_other is 0, PRR is infinite (strong signal)
        df_signals['PRR'] = np.where(
            df_signals['incidence_other'] > 0,
            df_signals['incidence_drug'] / df_signals['incidence_other'],
            np.inf
        )
        
        print("PRR calculation complete.\n")

        # --- Step 5: Identify and Display Potential Signals ---
        print("--- Step 5: Identifying Potential ADR Signals ---")
        # A common threshold for a "signal" is:
        # 1. PRR > 2 (The side effect is reported >2x as often for this drug)
        # 2. Count (a) >= 3 (The association is not just a random one-off report)
        
        potential_signals = df_signals[
            (df_signals['PRR'] > 2) & 
            (df_signals['a'] >= 3)
        ].sort_values(by='PRR', ascending=False)
        
        print(f"\nFound {len(potential_signals)} potential signals (PRR > 2 and Count >= 3).")
        print("\nTop 20 Potential Signals (Sorted by PRR):")
        
        # Select key columns for a cleaner output
        output_cols = ['name', 'Side Effect', 'Count', 'PRR', 'a_plus_c', 'a_plus_b']
        print(potential_signals[output_cols].head(20).to_string())
        

        # ========================================================
        # START: ADR SIGNAL DETECTION - VISUALIZATION CODE
        # ========================================================

        # Set style
        sns.set_style("whitegrid")
        plt.rcParams['figure.figsize'] = (14, 8)

        print("\n\nðŸŽ¨ Creating Professional Visualizations...\n")

        # ========================================
        # VISUALIZATION 1: Top 20 Side Effects (Bar Chart)
        # ========================================

        top_effects = df_melted['Side Effect'].value_counts().head(20)

        plt.figure(figsize=(12, 8))
        colors = plt.cm.Reds(np.linspace(0.4, 0.8, len(top_effects)))
        top_effects.plot(kind='barh', color=colors)
        plt.title('Top 20 Most Reported Side Effects', fontsize=18, fontweight='bold', pad=20)
        plt.xlabel('Number of Reports', fontsize=14, fontweight='bold')
        plt.ylabel('Side Effect', fontsize=14, fontweight='bold')
        plt.gca().invert_yaxis() # Show highest at the top
        plt.tight_layout()
        plt.savefig('top_side_effects.png', dpi=300, bbox_inches='tight')
        print("âœ… Saved: top_side_effects.png")
        # plt.show() # Disabled plt.show() to allow script to run fully

        # ========================================
        # VISUALIZATION 2: High-Risk Signals (PRR > 100)
        # ========================================

        # Filter for extremely high PRR signals
        high_risk = potential_signals[
            (potential_signals['PRR'] < np.inf) &
            (potential_signals['PRR'] > 100)
        ].head(15).sort_values('PRR') # Sort ascending for barh

        plt.figure(figsize=(14, 8))
        plt.barh(range(len(high_risk)), high_risk['PRR'], color='crimson')
        plt.yticks(range(len(high_risk)),
                   [f"{row['name'][:30]}\n({row['Side Effect'][:25]})"
                    for _, row in high_risk.iterrows()],
                   fontsize=10)
        plt.xlabel('Proportional Reporting Ratio (PRR)', fontsize=14, fontweight='bold')
        plt.title('Top 15 Highest Risk Drug-Side Effect Signals (PRR > 100)',
                  fontsize=16, fontweight='bold', pad=20)
        plt.axvline(x=2, color='orange', linestyle='--', linewidth=2, label='PRR = 2 (Signal Threshold)')
        plt.legend(fontsize=12)
        plt.tight_layout()
        plt.savefig('high_risk_signals.png', dpi=300, bbox_inches='tight')
        print("âœ… Saved: high_risk_signals.png")
        # plt.show()

        # ========================================
        # VISUALIZATION 3: Drug Safety Profile (Top 10 Drugs)
        # ========================================

        # Get top 10 drugs by total reports
        top_drugs = df_melted['name'].value_counts().head(10).index.tolist()

        # Count side effects for each drug
        drug_safety_data = []
        for drug in top_drugs:
            drug_effects = df_melted[df_melted['name'] == drug]['Side Effect'].value_counts()
            for effect, count in drug_effects.head(5).items():
                drug_safety_data.append({
                    'Drug': drug[:30],  # Truncate long names
                    'Side Effect': effect,
                    'Count': count
                })

        df_safety = pd.DataFrame(drug_safety_data)

        # Create heatmap-style visualization
        if not df_safety.empty:
            plt.figure(figsize=(14, 10))
            pivot_table = df_safety.pivot_table(
                values='Count',
                index='Drug',
                columns='Side Effect',
                fill_value=0
            )
            sns.heatmap(pivot_table, annot=True, fmt='g', cmap='YlOrRd',
                        cbar_kws={'label': 'Number of Reports'})
            plt.title('Safety Profile: Top 10 Drugs & Their Most Common Side Effects',
                      fontsize=16, fontweight='bold', pad=20)
            plt.xlabel('Side Effect', fontsize=12, fontweight='bold')
            plt.ylabel('Drug Name', fontsize=12, fontweight='bold')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.savefig('drug_safety_heatmap.png', dpi=300, bbox_inches='tight')
            print("âœ… Saved: drug_safety_heatmap.png")
            # plt.show()
        else:
            print("âš ï¸ Skipped: drug_safety_heatmap.png (no data for top drugs)")


        # ========================================
        # VISUALIZATION 4: Interactive Signal Explorer (Plotly)
        # ========================================

        # Prepare data for interactive chart
        interactive_data = potential_signals[
            (potential_signals['PRR'] < 1000) & (potential_signals['PRR'] > 0) &
            (potential_signals['Count'] >= 5)
        ].head(100)

        if not interactive_data.empty:
            fig = px.scatter(
                interactive_data,
                x='Count',
                y='PRR',
                size='a_plus_b',
                color='PRR',
                hover_data=['name', 'Side Effect'],
                title='Interactive ADR Signal Explorer (Hover for Details)',
                labels={'Count': 'Number of Reports', 'PRR': 'Signal Strength (PRR)'},
                color_continuous_scale='Reds'
            )

            fig.update_layout(
                height=700,
                font=dict(size=14),
                showlegend=True
            )

            fig.add_hline(y=2, line_dash="dash", line_color="orange",
                          annotation_text="Signal Threshold (PRR=2)")

            fig.write_html('interactive_signal_explorer.html')
            print("âœ… Saved: interactive_signal_explorer.html")
            # fig.show()
        else:
            print("âš ï¸ Skipped: interactive_signal_explorer.html (no data for plot)")


        # ========================================
        # VISUALIZATION 5: Signal Distribution
        # ========================================

        plt.figure(figsize=(12, 6))

        # Create bins for PRR ranges
        prr_bins = [0, 2, 5, 10, 50, 100, 1000, np.inf]
        prr_labels = ['<2 (No Signal)', '2-5', '5-10', '10-50', '50-100', '100-1000', '>1000']

        # Use all signals (not just 'potential_signals') for a full distribution
        # Also filter out 'inf' PRR for categorization
        all_signals_finite_prr = df_signals[
            (df_signals['PRR'] < np.inf) & (df_signals['PRR'] > 0)
        ].copy()
        
        if not all_signals_finite_prr.empty:
            all_signals_finite_prr['PRR_Category'] = pd.cut(
                all_signals_finite_prr['PRR'],
                bins=prr_bins,
                labels=prr_labels,
                right=False # Bins are [0, 2), [2, 5), etc.
            )

            category_counts = all_signals_finite_prr['PRR_Category'].value_counts().sort_index()

            colors_dist = ['grey', 'yellowgreen', 'orange', 'red', 'darkred', 'purple', 'black']
            category_counts.plot(kind='bar', color=colors_dist)
            plt.title('Distribution of Signal Strength (All Finite PRR Values)',
                      fontsize=16, fontweight='bold', pad=20)
            plt.xlabel('PRR Range', fontsize=14, fontweight='bold')
            plt.ylabel('Number of Signals', fontsize=14, fontweight='bold')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.savefig('signal_distribution.png', dpi=300, bbox_inches='tight')
            print("âœ… Saved: signal_distribution.png")
            # plt.show()
        else:
            print("âš ï¸ Skipped: signal_distribution.png (no data for distribution)")

        # ========================================
        # VISUALIZATION 6: Top Drugs by Signal Count
        # ========================================

        # Count how many signals each drug has
        drug_signal_counts = potential_signals.groupby('name').size().sort_values(ascending=True).tail(15) # .tail(15) for barh

        if not drug_signal_counts.empty:
            plt.figure(figsize=(12, 8))
            drug_signal_counts.plot(kind='barh', color='steelblue')
            plt.title('Top 15 Drugs by Number of ADR Signals Detected',
                      fontsize=16, fontweight='bold', pad=20)
            plt.xlabel('Number of Signals (PRR > 2)', fontsize=14, fontweight='bold')
            plt.ylabel('Drug Name', fontsize=14, fontweight='bold')
            plt.tight_layout()
            plt.savefig('top_drugs_by_signals.png', dpi=300, bbox_inches='tight')
            print("âœ… Saved: top_drugs_by_signals.png")
            # plt.show()
        else:
            print("âš ï¸ Skipped: top_drugs_by_signals.png (no drugs with signals)")

        print("\n" + "="*60)
        print("ðŸŽ‰ ALL VISUALIZATIONS COMPLETE!")
        print("="*60)
        print("\nðŸ“ Files Created:")
        print("1. top_side_effects.png")
        print("2. high_risk_signals.png")
        print("3. drug_safety_heatmap.png")
        print("4. interactive_signal_explorer.html")
        print("5. signal_distribution.png")
        print("6. top_drugs_by_signals.png")
        
        
        # --- ORIGINAL COMPLETION MESSAGE ---
        print("\n\n--- ADR Signal Detection Project Complete ---")
        print("Interpretation:")
        print(" - 'Count' (a): Number of reports for this specific drug-effect pair.")
        print(" - 'PRR': Proportional Reporting Ratio. A value of 5.0 means this side effect is reported 5x more frequently for this drug than for all other drugs in the dataset.")
        print(" - 'a_plus_c': Total side effect reports for this drug.")
        print(" - 'a_plus_b': Total reports for this side effect across all drugs.")

    except FileNotFoundError:
        print(f"ERROR: The file 'medicine_dataset.csv' was not found.")
        print("Please make sure the file is in the same directory as the script.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    run_adr_signal_detection()

```
